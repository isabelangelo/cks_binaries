```python
# now I just need to iterate throught these.
order_ranges = [(4976, 5056),
(5056, 5128),
(5128, 5203),
(5203, 5280),
(5280, 5360),
(5360, 5442),
(5442, 5526),
(5526, 5613),
(5613, 5703),
(5703, 5796),
(5796, 5891),
(5891, 5990),
(5990, 6093),
(6093, 6199),
(6199, 6309),
(6309, 6421)]

for order in range(16):
    order_range = order_ranges[order]
    wavedec_idx = (w>order_range[0]) & (w<order_range[1]) 
    w_order = w[wavedec_idx]
    flux_order = flux_koi1[wavedec_idx]
    
    if (len(w_order) % 2) != 0:
        w_order = w_order[:-1]
        flux_order =flux_order[:-1]

    plot_flux_waverec_levels(
        w_order, 
        flux_order, 
        'sym5', 
        'KOI-1, order {}'.format(order+1))
```
