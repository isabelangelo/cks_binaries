```python
%pylab inline
import pandas as pd
df = pd.read_csv('./data/cannon_models/cannon_order_model_stats.csv')
```

```python
n=1
for label in ['steff','slogg','smet','svsini']:
    subplot(4,1,n)
    df_label = df[df.label==label]
    plot(df_label.order, df_label.rms, 'bo')
    plot(df_label.order, df_label.rms, 'b--')
    ylim(0.7*df_label.rms.min(), 1.3*df_label.rms.max())
    ylabel(label[1:])
    #yscale('log')
    if n==4:
        xticks([1,2,3,4,5])
    else:
        xticks([])
    n+=1
xlabel('order number')
suptitle('CKS - Cannon rms')
```
