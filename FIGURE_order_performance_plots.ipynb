```python
# replace order column
import pandas as pd
#df_old = pd.read_csv('./data/cannon_models/rchip_order_stats_pre_mask.csv')
df = pd.read_csv('./data/cannon_models/rchip_order_stats.csv')

df_dwt = df[df.model.str.contains('dwt')]
df_original = df[df.model.str.contains('original')]
```
