```python
# next step: mask out region with telluric features
# these should be the same at every wavelength since the spectrum isn't registered
# Yee et al. masked out lines (see table for all of them)
# but there were much less telluric lines in that region.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
telluric_tbl = pd.read_csv('./telluric_atmabs.txt',skiprows=2, delim_whitespace=True)
plt.figure(figsize=(15,10))

plt.subplot(211)
for idx,row in telluric_tbl.iterrows():
    x = [row.Wavelength_min, row.Wavelength_max]
    y = row.Residual_Intensity*np.ones(len(x))
    plt.plot(x,y, lw=4)
plt.xlim(4975,6420);plt.title('r chip', fontsize=20)
plt.axvspan(6270,6310,color='r',alpha=0.3)
plt.text(6100,0.6,'masked out \nin Yee 2018', color='r',alpha=0.4)
plt.axhline(0.9,c='r', ls='--')
#plt.xlabel('wavelength (angstrom)')
plt.ylabel('Residual Intensity')

plt.subplot(212)
for idx,row in telluric_tbl.iterrows():
    x = [row.Wavelength_min, row.Wavelength_max]
    y = row.Residual_Intensity*np.ones(len(x))
    plt.plot(x,y, lw=4)
plt.xlim(6540,8000);plt.title('i chip', fontsize=20)
plt.axhline(0.9,c='r', ls='--')
plt.axvspan(6860,7030,color='lightgrey',alpha=0.3)
plt.axvspan(7170,7310,color='lightgrey',alpha=0.3)
plt.axvspan(7590,7700,color='lightgrey',alpha=0.3)
plt.xlabel('wavelength (angstrom)')
plt.ylabel('Residual Intensity')
```
